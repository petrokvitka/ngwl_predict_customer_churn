```python
test_shipments = test_shipments[test_shipments['s.order_state'] !='cart']
```

```python
test_shipments['s.order_state'].value_counts()
```

Категоризируем статус доставки на "доставлено" и "не доставлено", заказы с измененным составом отнесем к доставленным.

```python
def bool_group(order_state):
    if order_state == 'complete':
        return 1
    elif order_state == 'resumed':
        return 1
    else:
        return 0
test_shipments['is_complited'] = test_shipments['s.order_state'].apply(bool_group)
```

```python
test_shipments['is_complited'].value_counts()
```

```python
test_shipments['ship_time'] = (
    pd.to_datetime(test_shipments['shipped_at'])
    - pd.to_datetime(test_shipments['order_completed_at'])
)
```

Заполним пропуски во времени доставки нулями.

```python
test_shipments['ship_time'] = test_shipments['ship_time'].fillna(0)
```

Округлим время до часов и переведем в формат int.

```python
test_shipments['ship_time'] = (test_shipments['ship_time'].dt.round('1H') / np.timedelta64(1, 'h')).astype(int) 
```

```python
shipment_table = test_shipments[['phone_id','user_id','month', 
                            'is_complited', 'ship_time', 'total_weight', 
                            'total_cost', 'promo_total', 'rate', 'order_id', 'target']]
```

```python
shipment_table
```

Поделим данные на тестовую выборку и обучающую таким образом:

## Обучение моделей

```python
log_regr = LogisticRegression()
def train_ (model, x, y):
    x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size = 0.3, random_state = 42, shuffle=False)
    model = model
    model.fit(x_train, y_train)
    predict = model.predict(x_valid)
    f1 = f1_score(y_valid, predict)
    print(f1)   
```

```python
train_(log_regr, 
       train.drop(['phone_id','user_id', 'order_id', 'target'], axis=1),
       train['target'])
```
